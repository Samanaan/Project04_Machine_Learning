## Data ETL

```python
import pandas as pd
```

```python
retail_df = pd.read_csv("Resources/retail_price.csv")

retail_df.head()
```

```python
retail_df.info()
```

```python
retail_df.isnull().sum()
```

```python
# Remove unnecessary columns
retail_df = retail_df.drop(columns=["product_name_lenght", "product_description_lenght", "product_photos_qty", "product_weight_g", "weekday", "weekend", "holiday"])

retail_df.columns
```

```python
retail_df = retail_df.rename(columns={"qty":"qty_sold", "product_score":"product_rating", "customers":"no_customers", "s":"seasonality", "comp_1":"comp1_price", "ps1":"comp1_prod_rating", "fp1":"comp1_freight_price", "comp_2":"comp2_price", "ps2":"comp2_prod_rating", "fp2":"comp2_freight_price", "comp_3":"comp3_price", "ps3":"comp3_prod_rating", "fp3":"comp3_freight_price"})

retail_df.head()
```

### column_name_explaination: 
"seasonality": fluctuations in sales revenue caused by external factors and occur on a predictable schedule around the same time(s) every year
"comp1_price": product price of competitor 1; 
"comp1_prod_rating": product rating of competitor 1; 
"comp1_freight_price": product freight price of competitor 1
"lag_price": previous-month price of the product
