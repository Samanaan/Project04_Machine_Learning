### Data Cleaning

```python
# Replace all the values in the column 'ratings', which contain symbols other than 0-9 or '.',  with empty space 
retail_raw_df['ratings'] = retail_raw_df['ratings'].str.replace('[^0-9.]', '', regex=True)
```

```python
# Drop all null values
retail_raw_df = retail_raw_df.dropna()
```

```python
# Remove all non-digit symbols (currency symbol, comma) from the price columns 
retail_raw_df['discount_price'] = retail_raw_df['discount_price'].str.replace('[^0-9]', '', regex=True)
retail_raw_df['actual_price'] = retail_raw_df['actual_price'].str.replace('[^0-9]', '', regex=True)
retail_raw_df['no_of_ratings'] = retail_raw_df['no_of_ratings'].str.replace('[^0-9]', '', regex=True)
retail_raw_df
```

```python
# Drop all null values
retail_raw_df = retail_raw_df.dropna()
```

```python
# Get a brief summary of the retail_raw_df DataFrame.
retail_raw_df.info()
```

```python
# Replace other than digits or decimal point values with empty string
retail_raw_df['ratings'] = retail_raw_df['ratings'].str.replace('[^0-9.]', '', regex=True)
# retail_raw_df['ratings'] = retail_raw_df['ratings'].str.strip()

# Remove rows with empty ratings
retail_raw_df = retail_raw_df.loc[retail_raw_df['ratings'] != '', :]
```

```python
# Convert the'ratings', no_of_ratings', 'discount_price' and 'actual_price' columns to a numeric data type
retail_raw_df[['ratings', 'discount_price', 'actual_price']] = retail_raw_df[['ratings', 'discount_price', 'actual_price']].astype(float)
retail_raw_df
```

```python
retail_df = retail_raw_df
```

### Connecting to a Database

```python
# Create connection to SQLight database
connection = sqlite3.connect('retail.db')
```

```python
# Create 'retail_tb' table in 'retail_db' database to store 'retail_df' dataFrame
retail_df.to_sql('retail_tb', connection, index=False, if_exists='replace')
```

```python
# Create a cursor
cursor = connection.cursor()
```

```python
# Execute a query
query = "SELECT * FROM retail_tb WHERE ratings = 5.0"
cursor.execute(query)
```

```python
# Fetch the results and print
rows = cursor.fetchall()

for row in rows:
    print(row)
    print()
```

```python
# Close the cursor and connction
cursor.close()
connection.close()
```
